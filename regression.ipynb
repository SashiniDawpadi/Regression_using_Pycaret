```python
from pycaret.datasets import get_data
```

```python
df = get_data('insurance')
```

```python
df.info()
```

```python
df.describe()
```

```python
df.describe(include='all')
```

```python
from pycaret.regression import *
```

```python
reg = setup(df, target='charges', fold_strategy='kfold', fold=10)  # Use fold_strategy
check_model = compare_models(exclude=['catboost'], fold=4)
```

```python
GradientBoosting = create_model('gbr')
```

```python
plot_model(GradientBoosting)
```

```python
plot_model(GradientBoosting, plot='learning')
```

```python
plot_model(GradientBoosting, plot='feature')
```

```python
LightGBM = create_model('lightgbm')
```

```python
plot_model(LightGBM)
```

```python
tune_gbr = tune_model(GradientBoosting, n_iter=5,optimize="MSE")
```

```python
ensemble_gbr = ensemble_model(GradientBoosting,method='Bagging')
```

```python
stacked_model = stack_models([GradientBoosting, LightGBM])
```

```python
blended_model = blend_models([GradientBoosting, LightGBM])
```
